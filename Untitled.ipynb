```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LassoCV
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.shape
```

```python
np.unique(df['ID'].value_counts())
```

Выделить все бинарные признаки.

```python
bin_feature = []
for col in df.columns:
    if set((1,0)) == set(df[col].unique()):
        bin_feature.append(col)
        
bin_feature
```

```python
len(bin_feature)
```

Выделить категориальные признаки

```python
cat_feature = df.select_dtypes(include=[np.int64]).columns
len(cat_feature)
```

```python
train = df.drop(columns='ID')
train.head()
```

```python
train_cat_ohe =  pd.concat([pd.get_dummies(train[col]) for col in cat_feature[1:]], axis=1, keys=cat_feature[1:]).values
y = train['Score']
train_cat_ohe.shape
```

```python
train['col3802'].value_counts()
```

```python
num_feature = df.select_dtypes(include=[np.float64]).columns
sc = StandardScaler()
train_num_cat_scaler = sc.fit_transform(train[num_feature])
train_num_cat_scaler.shape
```

```python
X = np.concatenate((train_num_cat_scaler, train_cat_ohe), axis=1)
X.shape
```

```python
%%time
lr = LassoCV(cv=5).fit(X,y)
```

```python
lr.mse_path_ 
```

```python
[coef for coef in lr.coef_ if coef != 0]
```

```python
len(lr.coef_)
```

```python
from sklearn.feature_selection import SelectKBest
skb = SelectKBest(k=1000)
X_select_K = skb.fit_transform(X,y)
```
